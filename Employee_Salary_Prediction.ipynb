**Model Training**

A logistic regression model is trained using the cleaned and encoded data. We fit the model on the training set to learn patterns in the data.

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_scaled, y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print("Accuracy: ",accuracy_score(y_test, y_pred))
print("Confusion Matrix: ",confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

**Creating the User Interface**

A simple form-based interface using Gradio to collect input values and show predictions.

```python
!pip install gradio
```

```python
import gradio as gr
```

```python
def predict_income(age, education_num, capital_gain, capital_loss, hours_per_week,
                   workclass, marital_status, occupation, relationship, race, native_country, gender):

    gender_num = 1 if gender == "Male" else 0

    input_dict = {
        'age': [age],
        'educational-num': [education_num],
        'capital-gain': [capital_gain],
        'capital-loss': [capital_loss],
        'hours-per-week': [hours_per_week],
        'gender': [gender_num],
        'workclass_' + workclass: [1],
        'marital-status_' + marital_status: [1],
        'occupation_' + occupation: [1],
        'relationship_' + relationship: [1],
        'race_' + race: [1],
        'native-country_' + native_country: [1]
    }

    input_df = pd.DataFrame(columns=x.columns)
    input_df.loc[0] = 0

    for key in input_dict:
        if key in input_df.columns:
            input_df.at[0, key] = input_dict[key][0]

    input_scaled = scaler.transform(input_df)
    prediction = model.predict(input_scaled)[0]
    return "Income >50K" if prediction == 1 else "Income <=50K"
```

```python
with gr.Blocks() as interface:
    gr.Markdown("##  Employee Income Prediction", elem_id="title")

    with gr.Row():
        with gr.Column(scale=3):
            with gr.Row():
                age = gr.Number(label="Age")
                education_num = gr.Number(label="Education Num")
                hours_per_week = gr.Number(label="Hours/Week")

            with gr.Row():
                capital_gain = gr.Number(label="Capital Gain")
                capital_loss = gr.Number(label="Capital Loss")
                gender = gr.Radio(["Male", "Female"], label="Gender")

            with gr.Row():
                workclass = gr.Dropdown([col.replace("workclass_", "") for col in x.columns if col.startswith("workclass_")], label="Workclass")
                marital_status = gr.Dropdown([col.replace("marital-status_", "") for col in x.columns if col.startswith("marital-status_")], label="Marital Status")
                occupation = gr.Dropdown([col.replace("occupation_", "") for col in x.columns if col.startswith("occupation_")], label="Occupation")

            with gr.Row():
                relationship = gr.Dropdown([col.replace("relationship_", "") for col in x.columns if col.startswith("relationship_")], label="Relationship")
                race = gr.Dropdown([col.replace("race_", "") for col in x.columns if col.startswith("race_")], label="Race")
                native_country = gr.Dropdown([col.replace("native-country_", "") for col in x.columns if col.startswith("native-country_")], label="Native Country")

            submit_btn = gr.Button(" Predict Income")

        with gr.Column(scale=2):
            gr.Markdown("###  Prediction Result")
            output_text = gr.Textbox(label="Predicted Income", lines=2, interactive=False)

    submit_btn.click(
        fn=predict_income,
        inputs=[age, education_num, capital_gain, capital_loss, hours_per_week,
                workclass, marital_status, occupation, relationship, race, native_country, gender],
        outputs=output_text
    )
```
